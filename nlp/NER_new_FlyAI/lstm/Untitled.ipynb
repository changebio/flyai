```python
 def df2dict(df,words_dic,label_dic):
    w2l = {}
    for i in range(len(df)):
        question,answer = df.iloc[i]
        source=question.split()
        target = answer.split()
        for s,t in zip(source,target):
            if s in words_dic:
                print(s,t,words_dic[s],label_dic.index(t),len(w2l))
                w2l[words_dic[s]] = label_dic.index(t)
            else:
                print(s,t)
    return w2l
```
